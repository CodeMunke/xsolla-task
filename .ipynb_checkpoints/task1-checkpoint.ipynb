```python
import pandas as pd
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate , train_test_split , StratifiedShuffleSplit
from sklearn import   linear_model, metrics
from sklearn.linear_model import SGDClassifier as SGD

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

from statistics import mean
```

```python
# Читаем из .csv-шки
train_directory = "data1/train.csv"

data = pd.read_csv(train_directory, header=0)

data['y'] = [1 if y == 'yes' else 0 for y in data['y']]
```

# Начинаем отбор релевантных признаков
Одномерный отбор

```python
# Из датасета будут изъяты только численные неотрицательные признаки, для облегчения работы
X = data[['age', 'duration', 'campaign', 'pdays', 'previous', 'cons.price.idx', 'euribor3m', 'nr.employed']][:]
y = data['y'][:]

# Одномерный отбор признаков, по критерию хи-квадрат
bestfeatures = SelectKBest(score_func=chi2, k='all')
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Признак','Оценка']
print(featureScores.nlargest(5,'Оценка'))
```

## Отбор матрицей корреляций

```python
#Получим корреляции из всего датасета
corrmat = data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

### Можно воспользоваться встроенным полем feature_importances_, который есть в некоторых классификаторах

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt

model = ExtraTreesClassifier()
model.fit(X,y)
# Используем поле
print(model.feature_importances_)

# Рисуем график
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
# Данные признаки были выбраны, на основе всех 3-х тестов
# Удаление индекса потребителских цен увеличило точность
# Удаление "nr.employed" увеличило точность
# Удаление возраста незначительно уменьшило точность
X = data[['duration', 'pdays', 'previous', 'euribor3m']][:]
y = data['y'][:]
```

```python
# Создаем объект разделителя
# Random_state = const для воссоздоваемости результатов
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
```

```python
# Делим датасет на фолды
train_folds = []
test_folds = []

for train_idx, test_idx in sss.split(X, y):
    # print(train_idx)
    X_train = X.loc[train_idx, : ]
    # print(X_train)
    X_test = X.loc[test_idx, : ]
    # print(X_test)
    y_train = y.loc[train_idx]
    # print(y_train)
    y_test = y.loc[test_idx]
    
    train_folds.append((X_train, y_train))
    test_folds.append((X_test, y_test))
```

```python
# Обучение модели на фолдах
rfc = RandomForestClassifier(random_state=42)
auc_list = []
for t in range(0, 5):
    train_X = train_folds[t][0]
    train_y = train_folds[t][1]
    
    test_X = test_folds[t][0]
    test_y = test_folds[t][1]
    
    rfc.fit(train_X, train_y)
    pred = rfc.predict(test_X)
    auc = metrics.roc_auc_score(test_y, pred)
    print ('ROC AUC ',  t, ': ', auc)
    auc_list.append(auc)
```

```python
m = mean(auc_list)
print('Средний ROC AUC =', m)
```
